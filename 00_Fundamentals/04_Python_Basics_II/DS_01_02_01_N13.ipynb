## Ejercicio 1
Eleva todos los elementos de la lista al cuadrado
> ```ejer_1 = [1,2,3,4,5]```

`````python
def potencia2(list_num):
    for i in range(len(list_num)):
        list_num[i] = list_num[i]**2
    return list_num

ejer_1 = [1,2,3,4,5]
potencia2(ejer_1)
`````

## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_2 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
def duplicate(list_num, duplicate):
    list_indexes = []
    for i in range(len(list_num)):
        if list_num[i].lower() == duplicate.lower():
            list_indexes.append(i)
    return list_indexes

ejer_2 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

# Duplicados de "un"
duplicados_un = duplicate(ejer_2, "un")
print("Duplicados de 'un':", duplicados_un)

# Duplicados de "es"
duplicados_es = duplicate(ejer_2, "es")
print("Duplicados de 'es':", duplicados_es)

# Duplicados de "binario"
duplicados_binario = duplicate(ejer_2, "binario")
print("Duplicados de 'binario':", duplicados_binario)
`````

`````python
list_ejer3 = list(ejer_3)
print(list_ejer3, '-->', type(list_ejer3))

set_ejer3 = set(ejer_3)
print(set_ejer3, '-->', type(set_ejer3))

def tuple_to_dict(tpl):
    '''Convierte una tupla en un diccionario, usando los índices de la tupla como llaves en el diccionario '''
    dictionary = {}
    for index, value in enumerate(tpl):
        dictionary[index] = value
    return dictionary

dict_ejer3 = tuple_to_dict(ejer_3)
print(dict_ejer3, '-->', type(dict_ejer3))
`````

## Ejercicio 6
Multiplica todos los valores del diccionario
> ```ejer_6 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}```

`````python
def my_mult(dictionary):
    result = 1
    for value in dictionary.values():
        result *= value
    return result

ejer_6 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}

my_mult(ejer_6)
`````
