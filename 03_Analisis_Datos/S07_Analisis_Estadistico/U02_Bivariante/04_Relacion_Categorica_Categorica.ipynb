## Anlisis de Datos (Bivariante): Test de relacin entre variables categricas

Para ver si dos variables categricas presentan alguna relacin (cambios conjuntos, cambios en una de un tipo se dan al mismo tiempo que cambios de otro tipo en la otra, etc) hicimos el anlisis visual. Si mis datos son todos los datos que manejo, ese anlisis visual ya nos dice si existe la relacin, pero si mis datos no son todos o quiero extrapolarlos, tengo que ganar confianza en lo que he "visto" en el anlisis visual y para ello aplicamos los test.

### Test Chi-Cuadrado ()

Nosotros para ver la relacin entre dos variables categricas vamos a emplear el test , al final te dejo un anexo con las condiciones en las que se puede aplicar (ojo, no siempre podremos tenlo en cuenta), y el mtodo "manual" para hacer el test. Ahora vamos a ir al grano... 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from bootcampviztools import plot_categorical_numerical_relationship, plot_combined_graphs, \
    pinta_distribucion_categoricas, plot_grouped_boxplots, plot_categorical_relationship_fin

df_seguros = pd.read_csv("./data/Marketing-Customer-Analysis.csv")
df_air_jun = pd.read_csv("./data/dataset_viajes_jun23.csv")
```

#### Procedimiento

(Insisto si los datos son toda la poblacin y no vamos a extrapolar: no hay relacin entre response y estado, hay una relacin interesante entre response y el tipo de oferta)

**#1 Obtenemos la tabla de contingencia** (las frecuencias de aparicin de la combinacin de valores). Las columnas deben ser categricas.

```python
tabla_contingencia = pd.crosstab(df_seguros['response'], df_seguros['state'])
tabla_contingencia
```

**#2 Aplicamos la prueba** (utilizando la funcin adecuada del mdulo scipy.stats) y mostramos los resultados

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(tabla_contingencia)


print("Valor Chi-Cuadrado:", chi2)
print("P-Value:", p)
print("Grados de Libertad:", dof)
print("Tabla de Frecuencias Esperadas:\n", expected)
```

**#3 Interpretacin de resultados** (depende de lo que estemos esperando)

El test chi-2 tiene como hiptesis nula (o de partida) la independencia de las variables, por tanto un P-Value menor de 0.05 nos dira que podemos rechazar la hiptesis de partida con seguridad y pensar que existe una relacin estadsticamente significativa. 

Y en este caso? Pues en este caso, es todo lo contrario, refuerza la hiptesis de partida de que son independientes, aadiendo el anlisis visual... S, son independientes 

Aplicado de una vez a la otra relacin (response-renew_offer_type):

```python
tabla_contingencia = pd.crosstab(df_seguros['response'], df_seguros['renew_offer_type'])
chi2, p, dof, expected = chi2_contingency(tabla_contingencia)

print("Valor Chi-Cuadrado:", chi2)
print("P-Value:", p)
print("Grados de Libertad:", dof)
print("Tabla de Frecuencias Esperadas:\n", expected)
```

### ANEXO: Test Chi-cuadrado (): Cundo Aplicar la Prueba de Chi-Cuadrado?



- **Objetivo:** Se utiliza cuando quieres determinar si hay una asociacin significativa entre dos variables categricas. Por ejemplo, si quieres saber si hay una relacin entre gnero (hombre/mujer) y preferencia de un producto (me gusta/no me gusta).

- **Condiciones de Uso:** Idealmente, se utiliza cuando las categoras son mutuamente excluyentes y las frecuencias observadas en las categoras son suficientemente altas (generalmente, se espera que el 80% de las celdas tengan una frecuencia esperada de al menos 5).
