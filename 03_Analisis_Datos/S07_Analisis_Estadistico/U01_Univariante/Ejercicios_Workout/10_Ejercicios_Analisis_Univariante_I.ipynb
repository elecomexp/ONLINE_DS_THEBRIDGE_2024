### Ejercicio Extra #1:

Muestra las frecuencias absolutas y relativas de las variables categricas del dataset. Qu podras destacar?

```python
def cardinalidad(df_in, umbral_categoria, umbral_continua):
    df_out = pd.DataFrame([df_in.nunique(), df_in.nunique()/len(df_in) * 100, df_in.dtypes])
    df_out = df_out.T.rename(columns = {0: "Card", 1: "%_Card", 2: "Tipo"})
    

    df_out.loc[df_out["Card"] < umbral_categoria, "Clase"] = "Categrica"    
    df_out.loc[df_out["Card"] == 2, "Clase"] = "Binaria"
    df_out.loc[df_out["Card"] >= umbral_categoria, "Clase"] ="Numrica Discreta"
    df_out.loc[df_out["%_Card"] > umbral_continua, "Clase"] = "Numrica Continua"
    
    return df_out
```

```python
cardinalidad(df_tips, umbral_categoria=6, umbral_continua=30)
```

```python
columas_categricas = ['sex', 'smoker', 'day', 'time']
```

```python
for column in columas_categricas:
    print(f'Frecuencia absoluta "{column}": {df_tips[column].value_counts()}')
    BC.pinta_distribucion_categoricas(df_tips, [column])
    print(f'Frecuencia relativa "{column}": {df_tips[column].value_counts()/len(df_tips[column])}')
    BC.pinta_distribucion_categoricas(df_tips, [column], relativa=True)
    print('\n--------------\n')
```
