# Visualizacin Aplicada: Mostrando una Variable Numrica

Hagamos las importaciones necesarias y veamos que grficas y visualizaciones tenemos para hablar de una variable numrica. Cargamos datos y los limpiamos un poco, hoy tenemos nuevos amiguitos...

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


df_viajes = pd.read_csv("./data/dataset_viajes_jun23.csv")
df_seguros = pd.read_csv("./data/Marketing-Customer-Analysis.csv")
df_bitcoin = pd.read_csv("./data/BTC.csv")
```

```python
# Dataset Bitcoins
df_bitcoin['fechas'] = pd.to_datetime(df_bitcoin['date'])
df_bitcoin.set_index('fechas', inplace = True)
df_bitcoin.drop('date', axis = 1, inplace = True)
df_bitcoin.head()
```

### Distribucin: Histrogramas, Funciones Densidad (KDE), Combinado


Empezamos con el histograma, ese viejo compaero ya (tanto que no te hago escribir)

```python
fig, axs = plt.subplots(1, 1, figsize = (4.5,3))
sns.histplot(df_viajes["ingresos"]/1000,
             kde = False,
             color = 'r',
             bins = 100,
             ax = axs)      # Otra forma, sin recurrir al dataframe como argumento
axs.set_xlabel("Ingresos  (millones)")
axs.set_ylabel("Num_Vuelos");
```

Veamos ahora el Kernel Density Estimation (KDE) o nuestra estimacin de la funcin de densidad:

```python
fg = sns.displot(x='customer_lifetime_value', data=df_seguros, kind = 'kde', height = 4, aspect = 3)

# height controla la altura del grfico
# Aspect la relacin ente ancho/alto
# Ojo este grfico se puede asignar a un "axes" pero en realidad es para mostrar uno por "figure" (es slo por completitud)
```

Y ahora los dos juntos:

```python
fig, axs = plt.subplots(1, 1, figsize = (4.5,3))
sns.histplot(df_viajes["ingresos"]/1000,
             kde = True,
             color ='r',
             bins =100,
             ax = axs)  # Otra forma, sin recurrir al dataframe como argumento
axs.set_xlabel("Ingresos  (millones)")
axs.set_ylabel("Num_Vuelos");
```

### Distribucin (II): Caja, Enjambre, Violn

#### Caja

Nuestro querido y conocido diagrama de caja:




```python
fig, axs = plt.subplots(1, 1, figsize= (4.5,3))
sns.boxplot(x = "total_claim_amount", data = df_seguros, orient = 'h'); # Si no pongo nada se aade al ltimo
axs.set_xlabel("Reclamado  ($)")
axs.set_ylabel("Boxplot");
```

### Enjambre 

Este es nuevo y es interesante para anlisis, no s hasta que punto lo es para visualizacin salvo que en l se vea algo que no se pueda mostrar de otra forma. Este diagrama nos muestra los puntos 

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(x = 'consumo_kg',
              data = df_viajes,
              color = 'k')
```

Nos muestra para los valores de consumo los puntos sin solape en cada valor. Se suele complementar con un grfico de caja o de violn para ver la dispersin y la distribucin:

```python
# l solo los ajusta para que se superpongan

plt.figure(figsize=(8,6))
sns.boxplot(x = "consumo_kg", data = df_viajes)
sns.swarmplot(x = "consumo_kg", data=df_viajes,
             color="black");
```

Y para terminar esta parte, los violines (musica maestro):

```python
plt.figure(figsize = (20,5))
sns.violinplot(df_viajes['consumo_kg'], color = 'steelblue', orient = 'h')
```


### Evolucin: Lnea y rea

Para acabar la sesin veamos como pintar una variable cuando esta representa una secuencia de valores ordenados, como una serie temporal. El ejemplo que usamos son las cotizaciones diarias de bitcoin hasta 2022:

```python
# Tan sencillo como pintar un grfico de lnea, escogiendo bien el eje-x, por eso cambiamos a la fecha:
sns.lineplot(x = df_bitcoin.index, y = 'close', data = df_bitcoin)
```
